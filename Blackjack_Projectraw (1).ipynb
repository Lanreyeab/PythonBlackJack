### Rules:
`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>

**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.
 
`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust.

```python
import random
from random import shuffle
from IPython.display import clear_output

class Deck:
    
    def __init__(self):
        self.suits = ['Clubs', 'Diamonds', 'Hearts', 'Spades']
        self.ranks = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
        self.cards = []

    def get_deck(self):
        for suit in self.suits:
            for rank in self.ranks:
                self.cards.append((rank, suit))

    def shuffle_deck(self):
        random.shuffle(self.cards)
    
        
class Player:
    
    def __init__(self):
        self.hand = []
        self.score = 0

    def calculate_score(self):
        self.score = 0
        num_aces = 0

        for card in self.hand:
            rank = card[0]
            if rank.isnumeric():
                self.score += int(rank)
            elif rank in ['Jack', 'Queen', 'King']:
                self.score += 10
            elif rank == 'Ace':
                num_aces += 1

        # Adjust score for aces
        for _ in range(num_aces):
            if self.score + 11 <= 21:
                self.score += 11
            else:
                self.score += 1

    def bust(self):
        return self.score > 21

    def show_hand(self, show_all_cards=False):
        if show_all_cards:
            for card in self.hand:
                print(f'{card[0]} of {card[1]}')
        else:
            print(f'{self.hand[0][0]} of {self.hand[0][1]} and an unknown card')

    def has_blackjack(self):
        return self.score == 21 and len(self.hand) == 2

    def is_winner(self, dealer_score):
        return (not self.bust() and self.score > dealer_score) or (not self.bust() and dealer_score > 21)

class Human(Player): 
    
    def __init__(self):
        super().__init__()

    def display(self):
        print("You are the player.")
    

class Dealer(Player):
    
    def __init__(self):
        super().__init__()

    def display(self):
        print("You are the dealer.")

    def deal_hand(self, deck, player):
        player.hand.append(deck.cards.pop())
        player.hand.append(deck.cards.pop())

    def hit(self, deck, player):
        player.hand.append(deck.cards.pop())

    def reveal_hand(self):
        print("Dealer's hand:")
        for card in self.hand:
            print(f'{card[0]} of {card[1]}')
            
class Game:
    def __init__(self):
        self.dealer = Dealer()
        self.human = Human()

    def display_player_win(self):
        print("You win!")

    def display_push(self):
        print("Push! It's a tie.")

    def display_player_loss(self):
        print("You lose!")

    def main(self):
        num_decks = int(input("Enter the number of decks to use: "))
        print(f"Using {num_decks} deck(s).")

        deck = Deck()
        for _ in range(num_decks):
            deck.get_deck()
        deck.shuffle_deck()

        self.dealer.deal_hand(deck, self.human)
        self.dealer.deal_hand(deck, self.dealer)

        self.human.calculate_score()
        self.dealer.calculate_score()

        print("Your hand:")
        self.human.show_hand()
        if self.human.has_blackjack():
            print("You have Blackjack! You win!")
            return

        while not self.human.bust():
            choice = input("Do you want to hit? (y/n): ")
            if choice.lower() == 'y':
                self.dealer.hit(deck, self.human)
                self.human.calculate_score()
                self.human.show_hand()
            else:
                break

        if self.human.bust():
            print("You bust! You lose!")
            return

        self.dealer.reveal_hand()
        while self.dealer.score < 16:
            self.dealer.hit(deck, self.dealer)
            self.dealer.calculate_score()

        print("Dealer's final hand:")
        self.dealer.show_hand(show_all_cards=True)

        if self.dealer.bust() or self.human.is_winner(self.dealer.score):
            self.display_player_win()
        elif self.human.score == self.dealer.score:
            self.display_push()
        else:
            self.display_player_loss()


# Run the game
if __name__ == "__main__":
    game = Game()
    game.main()
```
